```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.read_csv('data/coupons.csv')

```

```python
def print_unique_values_for_object_columns(df):
    """
    Print unique values for each column that is of type 'object' in the DataFrame.
    
    Args:
        df (pandas.DataFrame): The DataFrame to analyze
    """
    # Get columns that are of object type (typically strings/categorical)
    object_columns = df.select_dtypes(include=['object']).columns
    
    print(f"Found {len(object_columns)} object-type columns:")
    print("=" * 60)
    
    for col in object_columns:
        unique_vals = df[col].unique()
        null_count = df[col].isnull().sum()
        
        print(f"\nColumn: '{col}'")
        print(f"Number of unique values: {len(unique_vals)}")
        print(f"Missing values: {null_count}")
        print(f"Unique values: {list(unique_vals)}")
        print("-" * 40)
    
    return object_columns

# Call the function to analyze the data
print_unique_values_for_object_columns(data)
```
